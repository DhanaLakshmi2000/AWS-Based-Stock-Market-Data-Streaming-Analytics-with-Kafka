```python
from kafka import KafkaConsumer
import json
from json import loads
from s3fs import S3FileSystem

# Initialize the Kafka Consumer
consumer = KafkaConsumer(
    'SM',  # Replace with your Kafka topic
    bootstrap_servers=['34.231.73.96'],  # Replace with your Kafka broker IP/hostname
    value_deserializer=lambda x: loads(x.decode('utf-8')),  # Deserialize the message
    auto_offset_reset='earliest',  # Optional: ensures you consume from the beginning
    enable_auto_commit=True  # Optional: commits offsets automatically
)

# Initialize the S3FileSystem (Make sure you have AWS credentials configured)
s3 = S3FileSystem()

# Start consuming messages from Kafka
for count, message in enumerate(consumer):
    try:
        print("Processing message number:", count)

        # Write the message to S3
        with s3.open(f"s3://aws-kafka-stock-market/stock_market_{count}.json", 'w') as file:
            json.dump(message.value, file)
            print(f"Message {count} written to S3 successfully.")
    
    except Exception as e:
        print(f"Error processing message {count}: {e}")

```
